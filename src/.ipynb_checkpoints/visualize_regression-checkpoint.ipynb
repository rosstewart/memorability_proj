```python
import matplotlib.pyplot as plt, numpy as np

preds, labels = [],[]
with open('test_out_resnet50_0.txt','r') as f:
    for line in f:
        id_, pred, label = line.split()
        preds.append(pred)
        labels.append(label)
preds, labels = np.array(preds).astype('float'), np.array(labels).astype('float')
```

```python
# plt.scatter(preds,labels)
# plt.axis('equal')
# plt.ylim([])
Y_test = labels
Y_test_pred = preds
fig, ax = plt.subplots(1,1,constrained_layout=True,figsize=(10,10))
ax.set_title('Y_test_pred vs Y_test memorability')
ax.scatter(Y_test, Y_test_pred,s=10)
ymax=np.max([Y_test.max(), Y_test_pred.max()])
ax.plot(np.linspace(0,ymax, 3), np.linspace(0, ymax, 3), '-r')
ax.set_xlabel('Y_test')
ax.set_ylabel('Y_test_pred')
ax.set_xlim([0.2,1])
ax.set_ylim([0.2,1])
```
